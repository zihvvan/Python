```python
from dateutil.relativedelta import relativedelta

start_date = datetime.now() - relativedelta(months=3)
history = yf.Ticker("TSLA").history(start=start_date)
history.head()
```

```python
before_month = lambda x: datetime.now() - relativedelta(months=x)
before_month(3)
stock_before_month = lambda x, y: yf.Ticker(x).history(start=before_month(y))
stock_before_month("AAPL", 6)
```
